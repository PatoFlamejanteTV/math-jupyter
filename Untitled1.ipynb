```python
# sin/cos/tan graphical demo using plot-like system

import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 2*pi
x = np.linspace(0, 2 * np.pi, 100)

# Calculate sin(x), cos(x), and tan(x)
sin_x = np.sin(x)
cos_x = np.cos(x)
tan_x = np.tan(x)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot sin(x)
plt.plot(x, sin_x, label='sin(x)', color='blue')

# Plot cos(x)
plt.plot(x, cos_x, label='cos(x)', color='red')

# Plot tan(x) - note the asymptotes
plt.plot(x, tan_x, label='tan(x)', color='green')

# Customize the plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('Trigonometric Functions')
plt.grid(True)
plt.legend()


# Adjust the y-axis limits for better visualization of tan(x)
plt.ylim(-5, 5)  # Adjust limits as needed

# Display the plot
plt.show()
```

```python
# asin/acos/atan graphical demo using plot-like system

import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-1, 1, 400)

# Calculate arcsin(x), arccos(x), and arctan(x)
arcsin_x = np.arcsin(x)
arccos_x = np.arccos(x)
arctan_x = np.arctan(x)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot arcsin(x)
plt.plot(x, arcsin_x, label='arcsin(x)', color='blue')

# Plot arccos(x)
plt.plot(x, arccos_x, label='arccos(x)', color='red')

# Plot arctan(x)
plt.plot(x, arctan_x, label='arctan(x)', color='green')

# Customize the plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('Inverse Trigonometric Functions')
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```
